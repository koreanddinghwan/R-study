### 문제1. 컴퓨터가 생각한 1 ~ 30 사이의 숫자를 맞치는 게임 프로그램을 작성하세요.
- 1 ~ 30 사이의 정수 중에서 랜덤하게 생성한 정수를 맞힐 때까지 무한 반복되도록 하세요.
- 0을 입력하면 정답을 맞히지 못해도 무한 반복이 종료되도록 하세요.
- 아래와 같은 결과가 나오도록 하세요.

```python
answer = 7 #random number, 랜덤모듈에서 가져온 값이 정수형이라고 가정

print('<< 컴퓨터가 생각한 1~30 숫자 맞히기 >>')

while True:
    while True:#입력검증
        userInput = int(input('숫자 입력(종료 0):'))
        if (userInput >= 0 and userInput <= 30):
            break
    if (userInput == 0): #종료조건
        break
    if (userInput == answer): #정답조건
        print('정답!!')
        break
    elif (userInput > answer): #더 낮을 숫자 입력
        print('더 작은 숫자 입력!')
    else:
        print('더 큰 숫자 입력!') #더 높은 숫자 입력
    
```

### 문제2. 월(month)을 입력받아 계절을 출력하는 프로그램을 작성하세요.
- 3, 4, 5월은 봄
- 6, 7, 8월은 여름
- 9, 10, 11월은 가을
- 12, 1, 2월은 겨울
- Dictionary를 활용하여 계절과 월을 저장하세요.
- for문과 if문을 활용하여 결과 list에 결과들을 추가하여 한번에 보여지도록 작성해보세요.
- 아래와 같은 결과가 나오도록 하세요.

결과 예시 1)
결과 예시 2)

```python
weatherDict = {
    '1' : '겨울',
    '2' : '겨울',
    '3' : '봄',
    '4' : '봄',
    '5' : '봄',
    '6' : '여름',
    '7' : '여름',
    '8' : '여름',
    '9' : '가을',
    '10' : '가을',
    '11' : '가을',
    '12' : '겨울'
}

weatherMonth = list(weatherDict.keys()) #계절월 딕셔너리에서 키값 가져오기

#문자열 입력
while (True):
    userInput = input('월 입력(공백으로 나눠서 입력하세요.): ')
    userInput = userInput.split(' ') #문자열을 공백기준 split해서 리스트로 저장
    length = 0 #길이확인 변수
    error = False #딕셔너리 키값에 있는 지 확인할 변수
    for i in userInput: #딕셔너리에 있는 키값인지 확인
        if i in weatherMonth: 
            length += 1
        else: #딕셔너리에 없는 값이 입력으로 들어온다.
            error = True #플래그를 False로 변경
            break
    
    if (error == True or length != 3): #flag가 False거나 길이가 3이 아니라면 다시 입력받음
        continue
    else:
        break

print('<< 월로 계절 찾아보기 >>')

result = [] #정답배열
for month in userInput:#유저가 입력한 월
    for i in weatherMonth:
        if month == i:
            result.append(weatherDict[i]) #정답 배열에 append

#결과 출력
for i in result:
    print(i, end=' ')#줄바꿈을 공백으로 치환해 출력
```
